## Fig.3 DCM Heatmaps

```python
import os
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
# Set data directory

data_dir = r'/Users/katerina/Desktop/DevCerebToM/figures/DCM_heatmaps'
```

## Passers

```python
# Read DCM results
passers = pd.read_excel(os.path.join(data_dir, 'children/pass/passers_DCM_NoCovariates.xls'))

# Set column names equal to values in row index position 0
passers.columns = passers.iloc[0]

# Remove first row from DataFrame
passers = passers[1:]

# Fixed connections
passers_fix = passers.iloc[1:9, 1:11]

# Dataframe
passers_df = passers_fix.set_index(passers_fix.columns[0], drop=True)
passers_df.index.name = None
passers_df = passers_df.apply(pd.to_numeric, errors='coerce')
passers_df
```

```python
# Plot the heatmap
hm = sn.heatmap(passers_df, cmap="RdGy_r", annot=False, fmt=".3f", cbar=True, vmin=-0.8, vmax=0.8)
fig = hm.get_figure()
```
